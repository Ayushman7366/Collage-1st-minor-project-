```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Create Sample Hostel Dataset**

**Explanation: **

*   Each room has a fixed capacity
*   The “Occupied” column shows how many students are currently staying
*   The data is stored in a Pandas DataFrame (table format)




```python
data = {
    "RoomID": ["R101","R102","R103","R104","R105","R106","R107","R108"],
    "Capacity": [4,4,3,2,4,3,2,4],
    "Occupied": [3,4,1,0,2,3,1,4]
}

df = pd.DataFrame(data)
df
```

**Calculate Vacancy**

**Explanation:**

*   Vacancy is calculated using the formula:**Vacant = Capacity − Occupied**
*   This shows how many beds are empty in each room




```python
df["Vacant"] = df["Capacity"] - df["Occupied"]
df
```

**Mean Occupancy Statistics (Using NumPy)**

**Explanation:**

*   Calculates the average number of occupied beds per room
*  Calculates the average number of vacant beds per room
*   NumPy is used for statistical calculations



```python
mean_occupied = np.mean(df["Occupied"])
mean_vacant = np.mean(df["Vacant"])

print("Mean Occupancy:", mean_occupied)
print("Mean Vacancy:", mean_vacant)
```

**Total Occupied vs Vacant Beds**

**Explanation:**

*   Finds the total number of occupied beds in the hostel
*   Finds the total number of vacant beds
*   These values are used for visualization





```python
total_occupied = df["Occupied"].sum()
total_vacant = df["Vacant"].sum()

total_occupied, total_vacant
```

**Pie Chart (Occupied vs Vacant)**

**Explanation:**

*   Displays a pie chart showing the proportion of occupied and vacant beds
*  Makes the analysis visually clear and easy to understand



```python
labels = ["Occupied Beds", "Vacant Beds"]
sizes = [total_occupied, total_vacant]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Hostel Occupancy Distribution")
plt.show()
```

**Vacancy Pattern Analysist**

**Explanation:**

*   Identifies rooms that are completely vacant
*   Helps analyze vacancy patterns in the hostel



```python
vacant_rooms = df[df["Occupied"] == 0]
vacant_rooms
```

```python
df["Occupancy_Percentage"] = (df["Occupied"] / df["Capacity"]) * 100
df
```
